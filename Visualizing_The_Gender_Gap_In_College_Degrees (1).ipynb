The US Department of Education released a dataset containing the percentage of bachelor's degrees granted to women from 1970 to 2012. The dataset I am using below was cleansed by Randal Olson, a data scientist at University of Pennsylvania.

All Degrees
Degrees have been categorised into 3 groups

STEM: Science, Technology, Engineering and

Mathematics

Liberal Arts
Other

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')

cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
fig =plt.figure(figsize=(18,9))
## Generate first row of line charts. STEM degrees.
for sp in range(0,6):
    #Create subplots
    ax = fig.add_subplot(3,6,sp+1)
    #Create a line chart for women and men(100-women)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    #sets the x-axis limits to range from 1968 to 2011
    ax.set_xlim(1968, 2011)
    #sets the y-axis limits to range from 0 to 100
    ax.set_ylim(0,100)
    #Sets ticks only at 1 and 100
    ax.set_yticks([0,100])
    #create a grey line at y=50
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3, linewidth=2)
    ax.set_title(stem_cats[sp])

    if sp == 0:
        ax.text(2002, 87, 'Women')
        ax.text(2005, 8, 'Men')
    elif sp == 5:
        ax.text(2005, 65, 'Men')
        ax.text(2001, 32, 'Women')


fig.subplots_adjust(hspace=1.0)

plt.show()
```

```python
fig=plt.figure(figsize=(18,9))

for sp in range(0, 5):
  counter=sp
  ax=fig.add_subplot(3,6,sp+1)
  ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[counter]], c=cb_dark_blue, label='Women', linewidth=3)
  ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[counter]], c=cb_orange, label='Men', linewidth=3)
  for key,spine in ax.spines.items():
      spine.set_visible(False)
  ax.set_xlim(1968, 2011)
  ax.set_ylim(0,100)
  ax.set_title(lib_arts_cats[counter])
  ax.tick_params(bottom="off", top="off", left="off", right="off")
  ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3, linewidth=2)
  ax.set_yticks([0,100])
  if counter == 0:
    ax.text(2002, 87, 'Women')
    ax.text(2005, 8, 'Men')

fig.subplots_adjust(hspace=1.0)

plt.show()
```

```python
fig=plt.figure(figsize=(18,9))

for sp in range(0, 6):

  counter=sp
  ax=fig.add_subplot(3,6,sp+1)
  ax.plot(women_degrees['Year'], women_degrees[other_cats[counter]], c=cb_dark_blue, label='Women', linewidth=3)
  ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[counter]], c=cb_orange, linewidth=3)
  for key,spine in ax.spines.items():
        spine.set_visible(False)
  ax.set_xlim(1968, 2011)
  ax.set_ylim(0,100)
  ax.set_title(other_cats[counter])
  ax.tick_params(bottom="off", top="off", left="off", right="off")
  ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3, linewidth=2)
  ax.set_yticks([0,100])

  if counter == 0:
      ax.text(2002, 70, 'Women')
      ax.text(2005, 25, 'Men')

  elif counter == 5:
      ax.text(2005, 65, 'Men')
      ax.text(2001, 29, 'Women')

fig.subplots_adjust(hspace=1.0)



plt.show()
```
